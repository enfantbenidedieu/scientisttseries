```python
# Load dataset
import pandas as pd

href = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-total-female-births.csv"
series = pd.read_csv(href, header=0, parse_dates=[0], index_col=0)
display(series.head())
```

```python
from scientisttseries.tsaplots import ggacf,ggpacf
```

```python
p = ggpacf(series,alpha=0.05,zero=False)
print(p)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
dta = sm.datasets.sunspots.load_pandas().data
dta.index = pd.Index(sm.tsa.datetools.dates_from_range('1700', '2008'))
del dta["YEAR"]
p = ggacf(dta.values.squeeze(), lags=40)
print(p)
```

```python
# 
import seaborn as sns
df = sns.load_dataset('flights')

X = pd.Series(df["passengers"].values,index=pd.date_range(start="1/1/1949",freq="M",periods=len(df)))
X.head()
```
